```python
import numpy as np
import pylab as py
import matplotlib.pyplot as plt
%matplotlib inline
```

# Multiple plots in one figure

```python
def y1(x):
    return 2 * x**2
```

```python
def y2(x):
    return x
```

```python
x= np.linspace(-10, 10, 100)

plt.figure(figsize=(20,10))  # sets the x,y size of the plot

plt.subplot(2,3,1)   # 2x3 table of plots, fill number 1 (counting from top left) 
plt.plot(x,y1(x))
plt.axis([-10,10,-10,10])

plt.subplot(2,3,2)  # fill number 2 (just right of #1)  
plt.plot(x,y2(x))
plt.axis([-10,10,-10,10])

plt.subplot(2,3,4)   # fill number 4 (just below #1)
plt.plot(x,0.01*y2(x)**3)
plt.axis([-10,10,-10,10])

plt.subplot(2,3,6)   # fill number 6 (right-bottom corner)
plt.plot(x,-0.01*y2(x)**3)
plt.axis([-10,10,-10,10])

plt.tight_layout()   # for better display (???)

plt.savefig('fig1.pdf')    # saves to file
```
