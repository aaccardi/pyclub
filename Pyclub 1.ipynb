```python
import sys,os
import pylab as py
import numpy as np
%matplotlib inline  
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting data points

x=[1,2,3,4,5,6]
y = [10,20,30,40,50,60]

plt.plot(x,y)  # joins the data points 
```

```python
plt.plot(x,y,'ro')     # just the points
```

```python
plt.plot(x,y,'ro--')   # points joined by dashed lines
```

```python
plt.plot(x,y,'ro',linestyle='dashed')   # alternative syntax
```

```python
# now we add labels and decorate the plot
plt.plot(x,y,'ro',linestyle='dashed')
plt.xlabel('independent var')
plt.ylabel('dependent var')
plt.title('This is the TITLE!')
```

```python
# We can also define the x and y ranges

plt.plot(x,y,'ro',linestyle='dashed')
plt.axis([0,7,0,70]);  # the semicolon suppresses teh output (needed also outside of ipython??)
```

```python
# Now 2 data sets in one plot, and we use data errors!

z = [3,20,40,15,45,30]

plt.plot(x,y,'ro',linestyle='dashed',label='Theory')
plt.errorbar(x,z,fmt='gx',linestyle='dotted',label='Exp.',xerr=0.2, yerr=5, ecolor='black')
# NOTE: plt.plot DOES NOT like the fmt= option: just gives the formatting string
#       BUT plt.errorbar REQUIRES fmt= in front of the string

plt.legend(loc='upper left')
plt.axis([0,7,0,70]);  # the semicolon suppresses teh output (needed also outside of ipython??)
```
